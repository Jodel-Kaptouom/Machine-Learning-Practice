```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
face.shape
```

```python
h = int(face.shape[0])
w = int(face.shape[1])
face_1= face[int(h/4):int(3*h/4), int(w/4):int(3*w/4)]
face_1
print (face_1)
plt.imshow(face_1, cmap=plt.cm.gray)
```

```python
face_1[face_1 > 127] = 255
face_1
```

```python
face_1[face_1 < 127 ] = 0
f1 = face_1 
f1
```

```python
plt.imshow(f1, cmap=plt.cm.gray)
```
