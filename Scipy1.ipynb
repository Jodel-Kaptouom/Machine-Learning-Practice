```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
x=np.linspace(0,10,10)
y=x**2
plt.scatter(x,y)
```

```python
f= interp1d(x,y, kind="linear")
new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x, y)
plt.scatter(new_x, result, c="r")
```

```python
x=np.linspace(0,10,10)
y=np.sin(x)
plt.scatter(x,y)
```

```python
f= interp1d(x,y, kind="cubic")
new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x, y)
plt.scatter(new_x, result, c="r")
```

```python
x=np.linspace(0,2,100)
y=1/3*x**3 - 3/5*x**2 + 2 + np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

```python
def f(x,a, b, c, d):
    return a*x**3+b*x**2+c*x+d
```

```python
from scipy import optimize 
params, param_cov = optimize.curve_fit(f, x, y) # ceci donne le parametre params=[a, b, c, d] et param_cov=[covariance]
```

```python
plt.scatter(x,y)
plt.plot(x, f(x,params[0], params[1], params[2], params[3]), c="g", lw=3 )
```

```python
def f (x): 
    return x**2 + 15*np.sin(x)
```

```python
x= np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
optimize.minimize(f, x0=-8)   # x0 le point a partir duquel l'algo de minimization prend effet
```

x: [-6.738e+00] ceci c'est le minimum local nous voulons le minimum global en changant le x0

```python
m0 = -5
result = optimize.minimize(f, x0=m0).x  
```

```python
plt.plot(x,f(x), lw=3, zorder = -1)
plt.scatter (result, f(result), s=100, c= "r", zorder=1)
plt.scatter (m0, f(m0), s=200, marker ="+", c="g", zorder = 1)
plt.show()
```

```python
def f (x):
    return np.sin(x[0]) +np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
x= np.linspace(-3, 3, 100)
y= np.linspace(-3, 3, 100)

x,y = np.meshgrid(x,y)
plt.contour (x, y, f(np.array([x, y])), 20)

x0 = np.zeros((2, 1))
plt.scatter(x0[0], x0[1], marker = "+", c= "r", s=100)

result = optimize.minimize(f, x0=x0).x
plt.scatter (result[0], result[1], c="g", s=100)
print (result)
```
