```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,2,10)
y= x**2
print (x)
```

```python
plt.plot(x,y, c= "black", lw=3, ls= "--")
plt.show() #pour autre editeur de texte il faut l'ajouter
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y, label = "quadratique")
plt.plot(x,x**3, label = "cubique")
plt.title ("figure 1")
plt.xlabel("axe x")
plt.ylabel("axe y")
plt.legend()
plt.show()
plt.savefig("figure.png")
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title("graphique 1")
plt.subplot(2,1,2)
plt.plot(x,np.sin(x), label="sinus")
plt.plot(x, np.cos(x), label="cosinus")
plt.legend()
```

```python
dataset = { f"experience{i}": np.random.randn(100,3) for i in range(4)}
dataset["experience0"].shape
```

```python
def graphique (dataset): 
    for i in range (4):   
        plt.subplot(4,1,i+1)
        plt.plot(dataset[f"experience{i}"])
        plt.title ("experience {}".format(i))
        plt.show()
graphique (dataset)
```

```python
def graphique (data):
    n=len(data)
    plt.figure(figsize=(12, 30))
    for k,i in zip(data.keys(), range (1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title (k)
    plt.show()
graphique (dataset)
```

```python
f = lambda x,y: np.sin(x)+np.cos(x+y)
X= np.linspace(0,5, 100)
Y= np.linspace(0, 5, 100)
X,Y = np.meshgrid (X,Y)
Z = f(X,Y)
ax = plt.axes (projection = "3d")
ax.plot_surface(X, Y, Z, cmap="plasma")
```
