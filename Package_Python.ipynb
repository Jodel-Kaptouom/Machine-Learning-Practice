```python
def fibonacci (n):
    a,b = 0, 1
    resultat = []
    while a<n:
        print (a)
        resultat.append(a)
        a, b = b, a+b
    print (resultat)
    return resultat
liste = fibonacci(50)
```

```python
import statistics 
print (statistics.mean(liste))
print (statistics.variance(liste))
```
